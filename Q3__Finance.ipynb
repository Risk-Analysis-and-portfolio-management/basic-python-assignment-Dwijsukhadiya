```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime



def analyze_portfolio(tickers, weights, start_date, end_date):
  data = yf.download(tickers, start=start_date, end=end_date , interval='1d')

  for ticker in tickers:
    data[f'{ticker}_Daily_Return'] = data['Adj Close'][ticker].pct_change()

  data['portfolio_dailyreturn'] = 0

  for i in range(len(weights)):
    data['portfolio_dailyreturn'] += weights[i] * data[f'{tickers[i]}_Daily_Return']


  data['cumulative_returns'] = (1 + data['portfolio_dailyreturn']).cumprod()

  final_portfolio_value = 100000 * data['cumulative_returns'].iloc[-1]
  print("final value of portfolio is :",final_portfolio_value)

analyze_portfolio(
tickers=['AAPL','MSFT','GOOGL','TSLA','NVDA'],
weights=[0.3, 0.2, 0.2, 0.2, 0.1],
start_date='2023-01-01',
end_date='2023-12-31'
)
```
